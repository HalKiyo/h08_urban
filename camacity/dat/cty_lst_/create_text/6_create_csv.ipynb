# Degree of urbanization comparison

```python
# verification
dou_path = f'/mnt/c/Users/tsimk/Downloads/dotfiles/h08/camacity/dat/cty_lst_/dou_precision.txt'
columns = ['city_num', 'rate', 'ara_diff', 'pop_diff', 'true', 'gridsize']
df_dou = pd.read_csv(dou_path, sep='|', header=None, names=columns)
df_dou.head()
```

```python
rate_list = []
aradiff_list = []
popdiff_list = []

for i in range(1, 1861):
    # city_num に i が存在するかをチェック
    if i in df_dou['city_num'].values:
        rate_value = df_dou.loc[df_dou['city_num'] == i, 'rate'].values[0]
        aradiff_value = df_dou.loc[df_dou['city_num'] == i, 'ara_diff'].values[0]
        popdiff_value = df_dou.loc[df_dou['city_num'] == i, 'pop_diff'].values[0]
        rate_list.append(rate_value)
        aradiff_list.append(aradiff_value)
        popdiff_list.append(popdiff_value)
    else:
        rate_list.append('NA')
        aradiff_list.append('NA')
        popdiff_list.append('NA')
```

```python
df_fifth['rate_below_1500'] = rate_list
df_fifth['ara_below_1500'] = aradiff_list
df_fifth['pop_below_1500'] = popdiff_list
df_fifth.head()
```

```python
import numpy as np
average_dou = np.nanmean(popdiff_list)
print(average_dou)
ave_ara_dou = np.nanmean(aradiff_list)
print(ave_ara_dou)
```
