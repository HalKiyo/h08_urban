```python
import pandas as pd

#############
save_path = '/mnt/c/Users/tsimk/Downloads/dotfiles/h08/camacity/dat/cty_lst_/create_text/camacity_forth.txt'
#############

text_path = '/mnt/c/Users/tsimk/Downloads/dotfiles/h08/camacity/dat/cty_lst_/create_text/camacity_third.txt'
df_third = pd.read_csv(text_path, sep="|", header=None)
df_third.columns = ['city_number', 'status', 'type', 'wup_pop', 'full_pop', 
                  'clustered_pop', 'Country', 'grids', 'Region', 'latitude', 
                  'longitude', 'City Name', 'Continent']
df_third.head(5)
```

```python
regionnum_path = '/mnt/c/Users/tsimk/Downloads/research/H08/additional_data/WUP/lookuptable.xlsx'
df_regionnum = pd.read_excel(regionnum_path, sheet_name='Sheet1')
df_regionnum.head()
```

```python
df_merged = pd.merge(
    df_regionnum[['rank', 'city_code', 'city_name', '5region_code', '5region_name', '17region_code', '17region_name', '22region_code', '22region_name', 'country_code', 'country_name']],  # 必要なカラムだけ選択
    df_third[['city_number', 'status', 'type', 'wup_pop', 'full_pop', 'clustered_pop', 'grids', 'latitude', 'longitude']],  # 必要なカラムだけ選択
    left_on='rank', right_on='city_number',
    how='inner'
)
print(len(df_merged))
df_merged.head(5)
```

```python
df_final = df_merged.drop(columns=['city_number'])
df_final = df_final[['city_code', 'rank', 'status', 'type', 'latitude', 'longitude', 'city_name', 'wup_pop', 
    '5region_code', '5region_name', '17region_code', '17region_name', '22region_code', '22region_name', 'country_code', 'country_name',
    'full_pop', 'clustered_pop', 'grids']]
df_final.head(5)
```

```python
df_final.to_csv(save_path, sep='|', index=False, header=False)
```
